# Data Combining
Note: this notebook is only to combine the separate experiments files,But the separate csv file are available through its Author's publication link
https://data.mendeley.com/datasets/3fcrrdjjvk/5

## Import required libraries

```python
import os
import pandas as pd
```

## Defining Funcions

```python
def CombineCSVtoDataFrames(parent_dir, Engine_status):
    """
    Combine all .csv file in a parent dir into a singl pandas Data Frame, Taking 
    into concidration the subfolders name which indicates the fixed parameters in
    the experiment.

    Input args:
    - parent_dir: absolute or relative path to the parent foler of experiments
    """
    dfs = []
    file_counter = 0  # Counter for processed CSV files
    
    for root, dirs, files in os.walk(parent_dir):
        
        for file in files:
            if file.lower().endswith('.csv'):
                full_path = os.path.join(root, file)
                try:
                    df = pd.read_csv(full_path)
                    
                    # Extract Temperature value form folder name
                    temp_folder = os.path.basename(root)
                    temperature = int(temp_folder.split()[0])
                    
                    # Extract Humidity value form folder name
                    humidity_folder = os.path.basename(os.path.dirname(root))
                    humidity = int(humidity_folder.split('%')[0])
                    
                    # Extract RPM value form folder name
                    rpm_folder = os.path.basename(os.path.dirname(os.path.dirname(root)))
                    rpm = int(rpm_folder.split()[0])
                    
                    # Increment counter for each successfully read CSV
                    file_counter += 1
                    
                    # Add all columns
                    df['RPM'] = rpm
                    df['Humidity'] = humidity
                    df['Temperature'] = temperature
                    df['Engine_status'] = Engine_status
                    
                    dfs.append(df)
                    
                    print(f"Processed ({file_counter}): {full_path}")
                    
                except Exception as e:
                    print(f"Error processing {full_path}: {e}")
    
    if not dfs:
        print("No CSV files found or all files failed to process.")
        return pd.DataFrame()
    
    # Concatenate all DataFrames
    concatenated_df = pd.concat(dfs, ignore_index=True)
    
    print(f"\nTotal CSV files processed: {file_counter}")
    print(f"Final DataFrame shape: {concatenated_df.shape}")
    
    return concatenated_df
```

---

## Healthy Engine

```python
# Combining the Healthy Enging Measurements
Healthy_Enging_Measurements_parent_path = "data/Engine Journal Bearings Dataset/Healthy Bearings Dataset"
Healthy_Engine_status = 0
Healthy_Enging_df = CombineCSVtoDataFrames(Healthy_Enging_Measurements_parent_path,
                                           Healthy_Engine_status)
```
